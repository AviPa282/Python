```python
#1. Create tuple t1 with first 10 natural numbers and tuple t2 with first 10 odd numbers  Merge both the tuples into third tuple t3

t1 = ()

for i in range(1, 11):
  t1 += (i, )

print("Natural number list: ", t1)

t2 = ()

for i in range(1, 21, 2):
  t2 += (i, )

print("Odd number list: ", t2)

t3 = t1 + t2

print("Merged list: ", t3)
```

```python
#2. Create tuple t1 with first 10 natural numbers and tuple t2 with first 10 odd numbers  Create tuple t3 with first 5 from each of above tuples 

t1 = ()

for i in range(1, 11):
  t1 += (i, )

print("Natural number list: ", t1)

t2 = ()

for i in range(1, 21, 2):
  t2 += (i, )

print("Odd number list: ", t2)

store = t1[:5]
store2 = t2[:5]

t3 = store + store2

print("Merged tuple: ", t3)
```

```python
#4. Store first 20 prime numbers in tuple named prime 

prime = ()
count = 0
n = 1
while True:
  if count == 20:
    break
  flag = 0
  for i in range(2, n//2 + 1):
    if n % i == 0:
      flag = 1
      break
  if flag == 0:
    prime += (n, )
    count += 1
  n += 1

print("Prime number list: ", prime)
  
```

```python
#5. Display maximum element in a tuple input by user 

t = ()

elements = int(input("Enter how many elements in tuple: "))

for i in range(elements):
  num = int(input("Enter the value: "))
  t += (num, )

print("The tuple you entered: ", t)

max = t[0]
for num in t:
  if max < num:
    max = num

print("The maximum element in the tuple is: ", max)
```
