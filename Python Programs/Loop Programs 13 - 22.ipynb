```python
#Display Prime numbers between x and y given by the user  

x = int(input("Enter a number: "))
y = int(input("Enter a number: "))

for i in range(x, y+1):
  for j in range(1, i):
    if j % 2 == 1:
      print("A prime number is: ", j)
```

```python
#Keep reading inputs from user until he presses a zero. Display the number of positive and negative elements input by the user 
negatives = 0
positives = 0
num = int(input("Enter a number: "))
while num != 0:
  if num > 0:
    positives = positives + 1
  else:
    negatives = negatives + 1
  num = int(input("Enter a number: "))

print("Thee number of negative elements is: ", negatives)
print("Thee number of positive elements is: ", positives)



```

```python
range1_to_ten = 0
range10to100 = 0
rangegreater100 = 0
num = int(input("Enter a number: "))
while num != 0:
  if num > 0 and num <= 10:
    range1_to_ten = range1_to_ten + 1
  elif num <= 100:
    range10to100 = range10to100 + 1
  elif num > 100:
    rangegreater100 = rangegreater100 +1
  num = int(input("Enter a number: "))

print("Thee number of 1 to 10 elements is: ", range1_to_ten)
print("Thee number of 10 to 100 elements is: ", range10to100)
print("Thee number of greater to 100 elements is: ", rangegreater100)
```
