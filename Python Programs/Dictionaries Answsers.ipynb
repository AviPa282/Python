```python
class Employee:

  def __init__(self):
    self.userid = 0
    self.username = ""
    self.usersalary = 0

  def input_data(self):
    self.userid = int(input("Enter your userid: "))
    self.username = input("Enter your name: ")
    self.usersalary = int(input("Enter your salary: "))

  def gross_salary(self):
    if self.usersalary <= 20000:
      grossalary = self.usersalary - (20*self.usersalary/100) - (7*self.usersalary/100) - (8.33*self.usersalary/100)
    elif self.usersalary <= 50000 and self.usersalary > 20000:
       grossalary = self.usersalary - (30*self.usersalary/100) - (10*self.usersalary/100) - (8.33*self.usersalary/100)
    else:
      grossalary = self.usersalary - (35*self.usersalary/100) - (12*self.usersalary/100) - (8.33*self.usersalary/100)
    
    print("The gross salary of the Employee: ", grossalary)

  def display(self):
      print("Your id is: ", self.userid)
      print("Your name is: ", self.username)
      print("Your usersalary is: ", self.usersalary)

emp1 = Employee()
emp1.input_data()
emp1.gross_salary()
emp1.display()
```
